```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.integrate as integrate
import scipy.special as special
%matplotlib inline

omm = 0.3
omr = 0
oml = 0.7
H0 = 70
c = 300000

z = np.linspace(0., 1., 100)
y = H0*np.sqrt(omm*(1+z)**3+omr*(1+z)**4+oml)

plt.figure(figsize = (10,7))
plt.plot(z, y, linestyle='--' , linewidth=2., color='r', label=r'$H_0\sqrt{\Omega_M(1+z)^3+\Omega_R(1+z)^4+\Omega_\Lambda}$')
plt.xlabel('z')
plt.ylabel('H(z)')
plt.legend(fontsize=15, loc=2)
plt.show()
```

```python
from scipy.integrate import quad
def integrand(z):
     return c/(H0*(np.sqrt(omm*(1+z)**3+omr*(1+z)**4+oml)))

z_arr = np.linspace(0., 1., 100)    
#c = 300000
#b = 70
I1_arr = np.zeros(len(z_arr))

#plt.figure(figsize = (10,7))
#plt.plot(z_arr, I, linestyle='--' , linewidth=2., color='r', label=r'$Nyarlathotep$')
#plt.xlabel('x')
#plt.ylabel('y')
#plt.legend(fontsize=20, loc=2)
#plt.show()
#print(len(z_arr))
```

```python
for i in range(len(z_arr)):
    I1_arr[i] = np.array(quad(integrand, 0, z_arr[i])[0])

plt.figure(figsize = (10,7))
plt.plot(z_arr,I1_arr,linestyle='--' , linewidth=2., color='g', label='Comoving distance')
plt.xlabel('z')
plt.ylabel(r'$\chi(z)$')
plt.legend(fontsize=15, loc=2)
plt.show()
```

```python
I2_arr = np.zeros(len(z_arr))
for i in range(len(z_arr)):
    I2_arr[i] = (1+z_arr[i])*np.array(quad(integrand, 0, z_arr[i])[0])

plt.figure(figsize = (10,7))
plt.plot(z_arr,I2_arr,linestyle='--' , linewidth=2., color='r', label='Luminosity distance')
plt.xlabel('z')
plt.ylabel(r'$D_L(z)$')
plt.legend(fontsize=15, loc=2)
plt.show()
```

```python
I3_arr = np.zeros(len(z_arr))
for i in range(len(z_arr)):
    I3_arr[i] = (1/(1+z_arr[i]))*np.array(quad(integrand, 0, z_arr[i])[0])

plt.figure(figsize = (10,7))
plt.plot(z_arr,I3_arr,linestyle='--' , linewidth=2., color='b', label='Angular distance')
plt.xlabel('z')
plt.ylabel(r'$D_A(z)$')
plt.legend(fontsize=15, loc=2)
plt.show()
```

```python
plt.figure(figsize = (10,7))
plt.plot(z_arr, I1_arr, label='Comoving distance')
plt.plot(z_arr, I2_arr, label='Luminosity distance')
plt.plot(z_arr, I3_arr, label='Angular distance')
plt.legend(fontsize=15, loc=2)
plt.show()
```
